```python
lm_pos = np.genfromtxt('data/lmPosX.csv', delimiter=',')

lm1_meas = np.genfromtxt('data/measLM1PosX.csv', delimiter=',')
lm2_meas = np.genfromtxt('data/measLM2PosX.csv', delimiter=',')

start_loc = np.genfromtxt('data/robStartLocX.csv', delimiter=',')
gt_pos = np.genfromtxt('data/gtPosX.csv', delimiter=',')

times = np.arange(len(gt_pos))

plt.plot(times, lm1_meas, label='landmark 1 measured distance')
plt.plot(times, lm2_meas, label='landmark 2 measured distance')
plt.title('measured distances from landmarks')
plt.legend(loc='upper center')

plt.show()
```
